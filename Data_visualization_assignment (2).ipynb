5. Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a
heatmap.



Answer-

```python
import pandas as pd
#generate synthetic data with correlated features
np.random.seed(55)
#for reproducibility
n_samples = 200
data = {
    'feature1': np.random.rand(n_samples),
    'feature2': 2 * np.random.rand(n_samples) + 0.5 * np.random.rand(n_samples),
}
df = pd.DataFrame(data)
# calculate the correlation matrix
correlation_matrix = df.corr()
# visualize the correlation
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0)
plt.title('Correlation Matrix Heatmap')
plt.show()
```

4. Using the given x and y data, generate a 3D surface plot to visualize the function z = sin()

x = np.linspace(-5, 5, 100)

x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))

    'X': x.flatten(),
    'Y': y.flatten(),
    'Z': z.flatten()
}
df = pd.DataFrame(data)



Answer-

```python
import plotly.graph_objects as go

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))
data = {
    'X': x.flatten(),
    'Y': y.flatten(),
    'Z': z.flatten()
}
df = pd.DataFrame(data)

fig = go.Figure(data=[go.Surface(z=z, x=x, y=y)])
fig.update_layout(title='3D Surface Plot of sin(sqrt(x^2 + y^2))',
                  scene = dict(
                      xaxis_title='X',
                      yaxis_title='Y',
                      zaxis_title='Z'))
fig.show()
```

BOKEH ASSIGNMENT:

1.Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x.


Answer-

```python
from bokeh.plotting import figure, show, output_notebook
import numpy as np

# Output the plot in the notebook
output_notebook()

# Create x and y data
x = np.linspace(0, 10, 200)
y = np.sin(x)

# Create a new plot
p = figure(title="Sine Wave", x_axis_label="x", y_axis_label="sin(x)")

# Add a line renderer
p.line(x, y, line_width=2)

# Show the plot
show(p)
```

2.Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the
markers based on the 'sizes' and 'colors' columns.


Answer-

```python
from bokeh.plotting import figure, show, output_notebook
import numpy as np
import pandas as pd

# Output the plot in the notebook
output_notebook()

# Create random data
n = 50
x = np.random.rand(n)
y = np.random.rand(n)
sizes = np.random.randint(10, 20, n)
colors = ["#%02x%02x%02x" % (r, g, 150) for r, g in zip(np.random.randint(0, 255, n), np.random.randint(0, 255, n))]

# Create a DataFrame (optional, but useful for real-world data)
data = {'x': x, 'y': y, 'sizes': sizes, 'colors': colors}
df = pd.DataFrame(data)


# Create a new plot
p = figure(title="Scatter Plot with Sizes and Colors", x_axis_label="x", y_axis_label="y")

# Add a circle renderer with sizes and colors from the DataFrame
p.circle(x='x', y='y', size='sizes', color='colors', source=df, legend_field='sizes')

# Show the plot
show(p)
```
